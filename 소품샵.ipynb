```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium import plugins
df = pd.read_csv('/content/a.csv', encoding='cp949')

dataset = df[['상호명','지점명',
              '상권업종대분류코드','상권업종중분류명', '표준산업분류명',
              '시도명', '시군구명', '행정동명',
              '위도', '경도']]

plt.figure(figsize=(8, 4))
sns.scatterplot(x='경도',y='위도', data=dataset, alpha=0.5)
plt.title('위도경도 분포도')
plt.show()

plt.figure(figsize=(20, 4))
sns.countplot(x='시군구명', data=dataset, order=dataset['시군구명'].value_counts().index)
plt.xticks(rotation=90)
plt.title('시군구별 상호 수')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium import plugins

df = pd.read_csv('/content/a.csv', encoding='cp949')

dataset = df[['상호명','지점명',
              '상권업종대분류코드','상권업종중분류명', '표준산업분류명',
              '시도명', '시군구명', '행정동명',
              '위도', '경도']]

a_map = folium.Map(location=[dataset['위도'].mean(), dataset['경도'].mean()], zoom_start=12)

for index, row in dataset.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"{row['상호명']} - {row['시군구명']}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(a_map)

a_map
```
